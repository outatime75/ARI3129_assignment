```python
pizza_labels = pd.read_csv(r'/home/kcini75/Notebooks/ARI3129_Computer_vision/labels.csv')
```

```python
selected_toppings = ['image_name','pepperoni', 'mushrooms', 'peppers', 'black_olives']
```

```python
pizza_selections = pd.DataFrame(columns= selected_toppings)

for topping in ['pepperoni', 'mushrooms', 'peppers', 'black_olives']:
    
    # Creating subset of pizzas containing selected topping
    toppings_subset = pizza_labels[selected_toppings][pizza_labels[selected_toppings][topping] == 1].copy()
    
    # Select 100 random pizzas from the subset of the dataframe where selected_topping == 1
    selected_subset = toppings_subset.sample(100, random_state = 554)
    pizza_selections = pd.concat([pizza_selections,selected_subset], ignore_index=True)
```

```python
for t in ['pepperoni', 'mushrooms', 'peppers', 'black_olives']:
    print(f'Number of pizzas with {t} : {len(pizza_selections[pizza_selections[t] ==1])}')
```

```python
pizza_selections = pizza_selections.drop_duplicates()
```

```python
for t in ['pepperoni', 'mushrooms', 'peppers', 'black_olives']:
    print(f'Number of pizzas with {t} : {len(pizza_selections[pizza_selections[t] ==1])}')
```

```python
# Extracting names of selected pizzas

selected_filenames = [f for f in pizza_selections.image_name]
```

```python
# Using ChatGPT to draft code for selection of pizzas

import os
import shutil

# Replace these paths with your actual paths
original_folder = '/home/kcini75/Notebooks/ARI3129_Computer vision/original_pizza_files/pizza_data/images'
selections_folder = '/home/kcini75/Notebooks/ARI3129_Computer vision/selected_pizzas'

# Create the selections folder if it doesn't exist
if not os.path.exists(selections_folder):
    os.makedirs(selections_folder)

# Iterate over the desired filenames and copy the corresponding files
for filename in selected_filenames:
    source_path = os.path.join(original_folder, filename)
    destination_path = os.path.join(selections_folder, filename)

    # Check if the file exists before copying
    if os.path.exists(source_path):
        shutil.copyfile(source_path, destination_path)
        print(f"Copied {filename} to selections folder.")
    else:
        print(f"File {filename} not found in the original folder.")

print("Selection process completed.")
```
